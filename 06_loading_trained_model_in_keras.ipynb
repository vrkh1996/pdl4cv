<div style="direction:ltr;text-align:left;font-family: Tahoma">
Import the required libraries.<br>
</div>

```python
from keras.models import load_model
from keras.preprocessing import image
from matplotlib.pyplot import imshow
import numpy as np
%matplotlib inline
```

<div style="direction:ltr;text-align:left;font-family: Tahoma">
Setting the directory for images:<br>
</div>

```python
cat_image_path = "./images/cat.jpg"
dog_image_path = "./images/dog.jpg"
```

<div style="direction:ltr;text-align:left;font-family: Tahoma">
Loading and displaying images:
</div>

```python
cat_image = image.load_img(cat_image_path, target_size=(150, 150))
imshow(cat_image)
```

```python
dog_image = image.load_img(dog_image_path, target_size=(150, 150))
imshow(dog_image)
```

<div style="direction:ltr;text-align:left;font-family: Tahoma">
Preprocessing images:
</div>

```python
cat_image = image.img_to_array(cat_image)
cat_image /= 255.
cat_image = np.expand_dims(cat_image, axis=0)
```

```python
dog_image = image.img_to_array(dog_image)
dog_image /= 255.
dog_image = np.expand_dims(dog_image, axis=0)
```

<div style="direction:ltr;text-align:left;font-family: Tahoma">
Loading the model architecture and weights:
</div>

```python
model = load_model('models/cnn_catvsdog.h5')
```

<div style="direction:ltr;text-align:left;font-family: Tahoma">
Predicting with the model:
</div>

```python
print(model.predict(cat_image))
print(model.predict_classes(cat_image))
```

```python
print(model.predict(dog_image))
print(model.predict_classes(dog_image))
```
